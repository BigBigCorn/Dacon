- categorical feature 처리 중점

# lightgbm+가입일제거

```python
import lightgbm as lgb
```

```python
x_train_lgb = newtrain.drop(columns = ['가입일', '전화해지여부'])
x_test_lgb = test.drop(columns = ['ID', '가입일'])
y_train_lgb = newtrain['전화해지여부']
y_test_lgb = pd.read_csv('/content/drive/MyDrive/스터디/new_open.zip (Unzipped Files)/sample_submission.csv')['전화해지여부']
```

```python
clf = lgb.LGBMClassifier(learning_rate = 0.01,
    boosting_type = 'dart',
    max_depth = 100,
    num_leaves = 144,
    random_state = 1422)
clf.fit(x_train_lgb, y_train_lgb)
```

```python
y_pred = clf.predict(x_test_lgb); y_pred
```

```python
submit = pd.read_csv('/content/drive/MyDrive/스터디/new_open.zip (Unzipped Files)/sample_submission.csv')
submit['전화해지여부'] = y_pred
```

```python
submit.to_csv('/content/drive/MyDrive/스터디/new_open.zip (Unzipped Files)/submit2.csv', index = False)
```

```python
x_train_dt = newtrain.drop(columns = ['가입일', '전화해지여부'])
x_test_dt = test.drop(columns = ['ID','가입일'])
y_train_dt = newtrain['전화해지여부']
y_test_dt = pd.read_csv('/content/drive/MyDrive/스터디/new_open.zip (Unzipped Files)/sample_submission.csv')['전화해지여부']
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler()
```

```python
df_robust = scaler.fit_transform(x_train_dt)
```

```python
train_df = pd.DataFrame(df_robust); train_df
```

```python
test_df = pd.DataFrame(scaler.transform(x_test_dt)); test_df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 10, min_samples_split = 3, random_state = 1422)
```

```python
model.fit(train_df, y_train_dt)
```

```python
pred = model.predict(test_df)
```

```python
submit = pd.read_csv('/content/drive/MyDrive/스터디/new_open.zip (Unzipped Files)/sample_submission.csv')
submit['전화해지여부'] = pred
```
