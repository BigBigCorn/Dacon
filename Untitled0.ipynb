## Import

```python
import pandas as pd

# Tabular Data를 다룰 것이기 때문에 하단의 라이브러리 호출
from autogluon.tabular import TabularDataset, TabularPredictor
```

## Data Load

```python
train_df = pd.read_csv('/content/drive/MyDrive/2023 옥수수/data/전화해지여부분류/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/2023 옥수수/data/전화해지여부분류/test.csv')
```

```python
train_df.head()
```

## Train

```python
# autogluon 학습을 위한 데이터 형태로 변환

train = TabularDataset(train_df.drop(['ID'], axis = 1))
test = TabularDataset(test_df.drop(['ID'], axis = 1))
```

```python
# 하단처럼 한 줄만 작성하면 모델 내부에서 알아서 학습해줌
# 우리가 가진 데이터로 모델 학습습

predictor = TabularPredictor(label = '전화해지여부', eval_metric = 'f1_macro').fit(train)
```

```python
# 각 모델의 훈련 성능 평가

ld_board = predictor.leaderboard(train, silent = True)
ld_board
```

```python
# 예측하기
pred_y = predictor.predict(test)
```

```python
# 제출 파일 생성
submit = pd.DataFrame()

submit['ID'] = test_df['ID']
submit['전화해지여부'] = pred_y
```
